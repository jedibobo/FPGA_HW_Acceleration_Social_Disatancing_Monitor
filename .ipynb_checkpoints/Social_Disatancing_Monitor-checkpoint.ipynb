# some fixed parameters

```python
anchor_list = [10,13,16,30,33,23,30,61,62,45,59,119,116,90,156,198,373,326]
anchor_float = [float(x) for x in anchor_list]
anchors = np.array(anchor_float).reshape(-1, 2)

classes_path = "img/voc_classes.txt"
class_names = get_class(classes_path)

```

```python
Yolo_Node_Names={
KERNEL_CONV : "tf_yolov3",
CONV_INPUT_NODE : "conv2d_1_convolution",
CONV_OUTPUT_NODE1 : "conv2d_59_convolution",
CONV_OUTPUT_NODE2 : "conv2d_67_convolution",
CONV_OUTPUT_NODE3 : "conv2d_75_convolution"}
```

```python
num_classes = len(class_names)
hsv_tuples = [(1.0 * x / num_classes, 1., 1.) for x in range(num_classes)]
colors = list(map(lambda x: colorsys.hsv_to_rgb(*x), hsv_tuples))
colors = list(map(lambda x: 
                  (int(x[0] * 255), int(x[1] * 255), int(x[2] * 255)), 
                  colors))
random.seed(0)
random.shuffle(colors)
random.seed(None)
```

# get results and draw boxes

```python
def evaluate(yolo_outputs, image_shape, class_names, anchors):
    score_thresh = 0.4
    anchor_mask = [[6, 7, 8], [3, 4, 5], [0, 1, 2]]
    boxes = []
    box_scores = []
    input_shape = np.shape(yolo_outputs[0])[1 : 3]
    input_shape = np.array(input_shape)*32

    for i in range(len(yolo_outputs)):
        _boxes, _box_scores = boxes_and_scores(
            yolo_outputs[i], anchors[anchor_mask[i]], len(class_names), 
            input_shape, image_shape)
        boxes.append(_boxes)
        box_scores.append(_box_scores)
    boxes = np.concatenate(boxes, axis = 0)
    box_scores = np.concatenate(box_scores, axis = 0)

    mask = box_scores >= score_thresh
    boxes_ = []
    scores_ = []
    classes_ = []
    for c in range(len(class_names)):
        class_boxes_np = boxes[mask[:, c]]
        class_box_scores_np = box_scores[:, c]
        class_box_scores_np = class_box_scores_np[mask[:, c]]
        nms_index_np = nms_boxes(class_boxes_np, class_box_scores_np) 
        class_boxes_np = class_boxes_np[nms_index_np]
        class_box_scores_np = class_box_scores_np[nms_index_np]
        classes_np = np.ones_like(class_box_scores_np, dtype = np.int32) * c
        boxes_.append(class_boxes_np)
        scores_.append(class_box_scores_np)
        classes_.append(classes_np)
    boxes_ = np.concatenate(boxes_, axis = 0)
    scores_ = np.concatenate(scores_, axis = 0)
    classes_ = np.concatenate(classes_, axis = 0)

    return boxes_, scores_, classes_
```

```python
def draw_boxes(image, boxes, scores, classes,indexes):
    _, ax = plt.subplots(1)
    ax.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    image_h, image_w, _ = image.shape

    for i, bbox in enumerate(boxes):
        [top, left, bottom, right] = bbox
        width, height = right - left, bottom - top
        center_x, center_y = left + width*0.5, top + height*0.5
        score, class_index = scores[i], classes[i]
        label = '{}: {:.4f}'.format(class_names[class_index], score) 
        color = tuple([color/255 for color in colors[class_index]])
        if i in indexes:
            color=(1,0,0)
        ax.add_patch(Rectangle((left, top), width, height,
                               edgecolor=color, facecolor='none'))
        if classes[i]!=14:
            ax.annotate(str(i), (center_x, center_y), color=color, weight='bold', 
                    fontsize=12, ha='center', va='center')
    return ax
```

# safety detector

```python
def get_person_coordinates(boxes):
    person_coordinates=[]
    for i in range(len(boxes)):
        box = np.array(boxes[i])
        box = np.where(box < 0, 0, box)
        (top, left, bottom, right) = box
        label = str(classes[i])
        if label == '14':
            person_coordinates.append([left, top, right-left, bottom-top])
    return person_coordinates
```

```python
def distance_boxes (boxA, boxB):
    import math
    center_boxA = [(boxA[0] + boxA[2])/ 2.0, (boxA[1] + boxA[3])/2.0]
    center_boxB = [(boxB[0] + boxB[2])/ 2.0, (boxB[1] + boxB[3])/2.0]
    pixel_distance  = math.sqrt( ((center_boxA[0]-center_boxB[0])**2)+((center_boxA[1]-center_boxB[1])**2) )
    return pixel_distance

distance_threshold_ratio=0.022 #define distance threshold

def safety_detector(boxes,img):
    violate=[]
    for x in range(len(boxes)):
        for y in range(x+1,len(boxes)):
            if (distance_boxes(boxes[x],boxes[y])/img.shape[1])< distance_threshold_ratio:
                cx1,cy1=int(boxes[x][0]+0.5*boxes[x][2]),int(boxes[x][1]+0.5*boxes[x][3])
                cx2,cy2=int(boxes[y][0]+0.5*boxes[y][2]),int(boxes[y][1]+0.5*boxes[y][3])
                cv2.line(img, (cx1,cy1), (cx2,cy2), (0, 0, 255), 5)
                if not x in violate:
                    violate.append(x)
                if not y in violate:
                    violate.append(y)
                print([x,y])
    return violate
```

# DPU HW Acceleration

```python
# create DPU task
n2cube.dpuOpen()
kernel = n2cube.dpuLoadKernel(Yolo_Node_Names[KERNEL_CONV])
task = n2cube.dpuCreateTask(kernel, 0)
```

```python
def get_yolo_outputs(task,image_data,Yolo_Node_Names):
    #get yolo outputs
    input_len = n2cube.dpuGetInputTensorSize(task, Yolo_Node_Names[CONV_INPUT_NODE])
    n2cube.dpuSetInputTensorInHWCFP32(
        task, Yolo_Node_Names[CONV_INPUT_NODE], image_data, input_len)

    n2cube.dpuRunTask(task)

    conv_sbbox_size = n2cube.dpuGetOutputTensorSize(task, Yolo_Node_Names[CONV_OUTPUT_NODE1])
    conv_out1 = n2cube.dpuGetOutputTensorInHWCFP32(task, Yolo_Node_Names[CONV_OUTPUT_NODE1], 
                                                   conv_sbbox_size)
    conv_out1 = np.reshape(conv_out1, (1, 13, 13, 75))

    conv_mbbox_size = n2cube.dpuGetOutputTensorSize(task, Yolo_Node_Names[CONV_OUTPUT_NODE2])
    conv_out2 = n2cube.dpuGetOutputTensorInHWCFP32(task, Yolo_Node_Names[CONV_OUTPUT_NODE2], 
                                                   conv_mbbox_size)
    conv_out2 = np.reshape(conv_out2, (1, 26, 26, 75))

    conv_lbbox_size = n2cube.dpuGetOutputTensorSize(task, Yolo_Node_Names[CONV_OUTPUT_NODE3])
    conv_out3 = n2cube.dpuGetOutputTensorInHWCFP32(task, Yolo_Node_Names[CONV_OUTPUT_NODE3], 
                                                   conv_lbbox_size)
    conv_out3 = np.reshape(conv_out3, (1, 52, 52, 75))

    yolo_outputs = [conv_out1, conv_out2, conv_out3]  
    
    return yolo_outputs
```

# Config Display Port

```python
from pynq.lib.video import *
displayport = DisplayPort()

displayport.configure(VideoMode(640, 480, 24), PIXEL_RGB)
```

# set usb camera

```python
camera = cv2.VideoCapture(0)
# camera = cv2.VideoCapture("./TownCentreXVID_small.mp4")

width = 640
height = 480
camera.set(cv2.CAP_PROP_FRAME_WIDTH,width)
camera.set(cv2.CAP_PROP_FRAME_HEIGHT,height)
```

# webcam stream using imutils

```python
def draw_boxes_realtime(image, boxes, scores, classes,indexes):
    image_h, image_w, _ = image.shape
    for i, bbox in enumerate(boxes):
        [top, left, bottom, right] = bbox
        width, height = right - left, bottom - top
        center_x, center_y = left + width*0.5, top + height*0.5
        score, class_index = scores[i], classes[i]
        label = '{}: {:.4f}'.format(class_names[class_index], score) 
#         color = tuple([color/255 for color in colors[class_index]])
        if i in indexes:
            color=(0,0,255)
        else:
            color=(0,255,0)
        cv2.rectangle(image, (left, top), (left+width, top+height), color, 2)    
#         if classes[i]!=14:
#             ax.annotate(str(i), (center_x, center_y), color=color, weight='bold', 
#                     fontsize=12, ha='center', va='center')
#     return 
```

# run in real time

```python
number_frames = 50
start = time.time()

timeF = 2
c=1

for _ in range(number_frames):
    frame = displayport.newframe()
    camera.read(frame)
#     frame = resize(frame, width=800)
    image_size = frame.shape[:2]
    image_data = np.array(pre_process(frame, (416, 416)), dtype=np.float32)
    if(c%timeF == 0):
        yolo_outputs=get_yolo_outputs(task,image_data,Yolo_Node_Names)
        boxes, scores, classes = evaluate(yolo_outputs, image_size, 
                                          class_names, anchors)
        indexes= safety_detector(get_person_coordinates(boxes),frame)
    #print(indexes)
#     #print(type(frame))
    draw_boxes_realtime(frame, boxes, scores, classes,indexes)
    c = c + 1
    displayport.writeframe(frame)

end = time.time()
duration = end - start
print(f"Took {duration} seconds at {number_frames / duration} FPS")
```

```python
n2cube.dpuDestroyTask(task)
n2cube.dpuDestroyKernel(kernel)
```

```python
camera.release()
displayport.close()
```
